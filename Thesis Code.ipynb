```python
import zipfile
import os
import glob
import pandas as pd

# Specify the directory containing the zip files
zip_dir = r'C:\Users\theam\Documents\1. Academics\1. First Semester\1. ECON 199\3. Python Project\July Project\RTD Prices and Schedules'

# Extract all zip files in the directory and subdirectories
for root, dirs, files in os.walk(zip_dir):
    for file in files:
        if file.endswith('.zip'):
            zip_path = os.path.join(root, file)
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(zip_dir)

# After extraction, read all CSV files into DataFrames
csv_files = glob.glob(os.path.join(zip_dir, '**', '*.csv'), recursive=True)
dataframes = [pd.read_csv(f) for f in csv_files]
```

```python
# Concatenate all dataframes into one
if dataframes:
    july2025_mw = pd.concat(dataframes, ignore_index=True)
    display(july2025_mw)
else:
    print("No CSV files were loaded.")
```

```python
# Fill missing RUN_TIME values with TIME_INTERVAL if RUN_TIME is NaN and TIME_INTERVAL is valid
july2025_mw['RUN_TIME'] = pd.to_datetime(july2025_mw['RUN_TIME'], errors='coerce')
july2025_mw['TIME_INTERVAL'] = pd.to_datetime(july2025_mw['TIME_INTERVAL'], errors='coerce')

# If RUN_TIME is NaN but TIME_INTERVAL is valid, use TIME_INTERVAL's date and set time to 00:00
mask = july2025_mw['RUN_TIME'].isna() & july2025_mw['TIME_INTERVAL'].notna()
july2025_mw.loc[mask, 'RUN_TIME'] = july2025_mw.loc[mask, 'TIME_INTERVAL'].dt.normalize()

# Format as 'YYYY-MM-DD HH:MM'
july2025_mw['RUN_TIME'] = july2025_mw['RUN_TIME'].dt.strftime('%Y-%m-%d %H:%M')
july2025_mw['TIME_INTERVAL'] = july2025_mw['TIME_INTERVAL'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
# Convert RUN_TIME and TIME_INTERVAL in july2025_mp to military time (24-hour format)
# If RUN_TIME is just a date (e.g. '7/16/2025'), set it to '00:00' for that day
def fix_run_time(val, prev_time=None):
    try:
        dt = pd.to_datetime(val, errors='coerce')
        if pd.isnull(dt):
            return np.nan
        # If original string has no time, set to 00:00
        if len(str(val).strip()) <= 10:
            return dt.strftime('%Y-%m-%d 00:00')
        else:
            return dt.strftime('%Y-%m-%d %H:%M')
    except:
        return np.nan

july2025_mp['RUN_TIME'] = [fix_run_time(val) for val in july2025_mp['RUN_TIME']]
july2025_mp['TIME_INTERVAL'] = pd.to_datetime(july2025_mp['TIME_INTERVAL'], errors='coerce').dt.strftime('%Y-%m-%d %H:%M')

# Display to verify conversion
print(july2025_mp[['RUN_TIME', 'TIME_INTERVAL']].head())
```

```python
# Merge july2025_mw and july2025_mp on RUN_TIME, TIME_INTERVAL, and REGION_NAME, adding MARGINAL_PRICE from july2025_mp
july2025_combined = pd.merge(
    july2025_mw,
    july2025_mp[['RUN_TIME','TIME_INTERVAL','REGION_NAME','MARGINAL_PRICE']],
    on=['RUN_TIME','TIME_INTERVAL','REGION_NAME'],
    how='left',
    suffixes=('', '_mp')
)
display(july2025_combined)
```

```python
# Clean the merged DataFrame
import pandas as pd
from datetime import datetime
# Drop columns with all missing values
july2025_combined = july2025_combined.dropna(axis=1, how='all')
# Convert RUN_TIME to datetime for filtering
july2025_combined['RUN_TIME'] = pd.to_datetime(july2025_combined['RUN_TIME'], errors='coerce')
# Filter for dates from July 1, 2025 onwards
july2025_combined = july2025_combined[july2025_combined['RUN_TIME'] >= datetime(2025, 7, 1)]
# Remove rows with missing RUN_TIME (if any remain)
july2025_combined = july2025_combined[july2025_combined['RUN_TIME'].notna()]
# Display cleaned DataFrame
display(july2025_combined)
```

```python
july2025_filtered = july2025_combined[july2025_combined['SCHED_MW'] > 0]
display(july2025_filtered)
```
